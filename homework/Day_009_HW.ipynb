# [範例重點]
弱相關的相關矩陣 (Out[2]) 與散佈圖 (Out[3]) 之間的關係  
正相關的相關矩陣 (Out[4]) 與散佈圖 (Out[5]) 之間的關係

```python
#載入套件
import numpy as np
np.random.seed(1)
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

## 弱相關

```python
#隨機生成兩組1000個介於0~50的數的整數x,y,看看相關矩陣如何
x = np.random.randint(0,50,1000)
y = np.random.randint(0,50,1000)

#呼叫numpy裡的相關矩陣函數(corrcoef)
np.corrcoef(x,y)
```

```python
#將分布畫出來看看吧
plt.scatter(x,y)
```

## 正相關

```python
#隨機生成 1000 個介於 0~50 的數 x
x = np.random.randint(0,50,1000)

#這次讓y與x正相關，再增加一些雜訊
y = x + np.random.randint(0,10,1000)

#再次用numpy裡的函數來計算相關係數
np.corrcoef(x,y)
```

```python
#再看看正相關的x,y分布
plt.scatter(x,y)
```

# 作業

寫出負相關的變數, 並觀察相關矩陣以及分布圖

```python
#隨機生成 1000 個介於 0~50 的數 x
x = np.random.randint(0,50,1000)

#這次讓y與x負相關，再增加一些雜訊
y = x/-0.5 + np.random.normal(0, 10, 1000)

#再次用numpy裡的函數來計算相關係數
np.corrcoef(x,y)
```

```python
#將分布畫出來看看吧
plt.scatter(x,y)
```
